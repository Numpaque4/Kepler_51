### How to generate stellar masses and radius using truncated normals

Generate masses and radius according to observed uncertainties:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import truncnorm

# Solar density
rho_sun = 1400
```

```python
# For Xmean, Xstd, see: https://exoplanetarchive.ipac.caltech.edu/cgi-bin/TblView/nph-tblView?app=ExoTbls&config=keplerstellar&constraint=kepid+like+%2711773022%27
N = int(1e4)
Rmean = 0.943
Rstd = 0.05
Rs = truncnorm.rvs(a=-Rmean/Rstd, b=np.inf, loc=Rmean, scale=Rstd, size=N)
Mmean = 1.05
Mstd = 0.05
Ms = truncnorm.rvs(a=-Mmean/Mstd, b=np.inf, loc=Mmean, scale=Mstd, size=N)
rhos = rho_sun*Ms/Rs**3
```

Filter densities in the right range:

```python
rhomin = 900
rhomax = 3000
cond = (rhos>rhomin) & (rhos<rhomax)
rhos = rhos[cond]
len(rhos)
```

Histogram:

```python
plt.figure()
plt.hist(rhos,bins=100);
```
